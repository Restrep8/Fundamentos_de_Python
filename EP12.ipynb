```python
lista = ["A", "B", "b", "c", "E", "E", "f"]

eliminar_elemento = input(f"Digite el elemento que quiera eliminar de {lista}: ").lower()

nueva_lista = []
for elemento in lista:
    if elemento.lower() != eliminar_elemento:
        nueva_lista.append(elemento)

print(f"\nLista: {nueva_lista}")
```
